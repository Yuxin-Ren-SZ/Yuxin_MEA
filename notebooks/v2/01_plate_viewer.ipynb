# 01 — Plate viewer (redirect)

**The plate viewer is now a dashboard page, not a notebook or a pipeline task.** Phase 5 (commit history) removed it from `PipelineManager`'s DAG entirely — visualization shouldn't sit inside processing.


## How to view a plate now

```sh
yuxin-mea-dashboard --config pipeline_config.json
```

Open http://127.0.0.1:8050 and click **Plate viewer** in the navbar. Pick a recording from the dropdown, click **Load** — the figure renders inline. Click **Export HTML** to write a standalone `<figure_root>/<recording_key>/plate_viewer.html` (same format the old `PlateViewerTask` produced).

## Programmatic access (advanced)

If you need to render plates from a headless script, call the library directly:


````
from pathlib import Path

from yuxin_mea.analysis.plate_raster_synchrony import (
    PlateViewerConfig,
    build_plate_figure,
    load_plate_data,
    write_plate_viewer_html,
)

analysis_root = Path("./data/analysis")     # ← your analysis_root
recording_key = "SampleA/240415/PlateX/Network/001"

well_records = load_plate_data(
    burst_detection_root=analysis_root / "burst_detection_data",
    curation_output_root=analysis_root / "curation_data",
    recording_key=recording_key,
    rec_name="auto",
    experiment_cache_path=analysis_root / "experiment_cache.json",
)
fig = build_plate_figure(well_records, PlateViewerConfig(display_mode="both"))

# Either show inline (Jupyter) or write to disk:
# fig.show()
# write_plate_viewer_html(fig, Path("./plate_viewer.html"))
````
