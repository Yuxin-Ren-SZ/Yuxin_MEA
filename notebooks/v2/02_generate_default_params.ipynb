**This notebook has been largely replaced by the dashboard's Settings page.**

The Settings page generates and validates `pipeline_config.json` interactively — no JSON-template snapshot required.


## How to author a config now

```sh
yuxin-mea-dashboard --config pipeline_config.json
```

If the file doesn't exist, the dashboard launches in **config-only mode**: navigate to **Settings**, fill in the Globals tab (`data_root`, `analysis_root`, `figure_root`) and any per-task tabs, click **Save**. The file appears at the path you passed.


## If you still need a one-shot JSON snapshot

Useful for CI fixtures or a headless workflow. The example below registers every task with `ConfigManager` and writes a template — same behavior as the old notebook, but via the new namespace.


````
from pathlib import Path

from yuxin_mea.config import ConfigManager
from yuxin_mea.tasks import (
    PreprocessingTask, SortingTask, AutoMergeTask, AnalyzerTask,
    AutoCurationTask, BurstDetectionTask, IterativeBurstDetectionTask,
)

cm = ConfigManager()
cm.set_global("data_root", "/path/to/raw")
cm.set_global("analysis_root", "./analysis")
cm.set_global("figure_root", "./figures")
for tc in (
    PreprocessingTask, SortingTask, AutoMergeTask, AnalyzerTask,
    AutoCurationTask, BurstDetectionTask, IterativeBurstDetectionTask,
):
    cm.register_task(tc)

out = Path("pipeline_config.json")
if out.exists():
    print(f"{out} already exists — delete it first if you want a fresh template.")
else:
    cm.generate_template(out)
    print(f"wrote {out}")
````
