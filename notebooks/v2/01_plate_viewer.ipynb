# 01 — Plate viewer (redirect)

**This notebook has been replaced by a dashboard page.**

The `PlateViewerTask` output (raster + synchrony grid per recording) is now rendered interactively by the multipage dashboard.


## How to view a plate now

Run the dashboard:

```sh
yuxin-mea-dashboard --config pipeline_config.json
```

Open http://127.0.0.1:8050, click **Recordings** in the navbar, pick a row from the table, and the plate viewer renders inline.
